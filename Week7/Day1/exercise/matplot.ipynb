```python
sns.set_style('whitegrid')
titanic = sns.load_dataset('titanic')
titanic.head()
```

```python
sns.jointplot(x='fare', y='age', data=titanic)
```

```python
sns.displot(x='fare', data=titanic)
```

```python
sns.histplot(x='fare', data=titanic, bins=10)
```

```python
sns.boxplot(x='class', y='age',data=titanic)
```

```python
plt.figure(figsize=[10,5])
sns.swarmplot(x='class', y='age',data=titanic)
```

```python
sns.countplot(data=titanic, x='sex')
```

```python
ax = sns.countplot(data=titanic, x='sex')
for i in ax.patches:
    ax.annotate(f'\n{i.get_height()}', (i.get_x()+0.2, i.get_height()), ha='left', va='top', color='white', size=15)
```

```python
g = sns.FacetGrid(titanic, col='sex')
g.map(plt.hist,'age').add_legend()
```
